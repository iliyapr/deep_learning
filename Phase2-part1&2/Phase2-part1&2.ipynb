## Drawing Word Clouds:

### Creating "sense" column, based on the overall value:

```python
train_data_complete["sense"] = train_data_complete["overall"].apply(lambda x : "Positive" if x==5 or x==4 else "Negative" if x==1 or x==2 else "Neutral" if x==3 else "Null")
```

```python
train_data_complete["sense"].head()
```

```python
from wordcloud import WordCloud
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
# grouped_set = train_data_complete.groupby(["sense"])
```

```python
# grouped_set.head()
```

### Tokenizing Positive text:

```python
text_Positive = " ".join(train_data_complete.loc[train_data_complete["sense"]=="Positive"]['reviewText'])
text_Positive[:300]
```

```python
len(text_Positive)
```

```python
int(0.6*len(text_Positive))
```

```python
tokens_Positive = word_tokenize(text_Positive[:int(0.6*len(text_Positive))])
```

```python
del(text_Positive)
```

```python
stop_words = set(stopwords.words('english'))
```

### Positive Comments Word Cloud:

```python
tokens_Positive = [word for word in tokens_Positive if word.lower() not in stop_words]
wordcloud = WordCloud(width=2400, height=1200).generate(' '.join(tokens_Positive))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
del(tokens_Positive)
```

## Top 10 Products with most 5 star votes:

```python
grouped = train_data_complete.groupby("title")[["overall", "brand"]].value_counts().reset_index().rename(columns={0: "repetitions"})
grouped
```

```python
Top_10_Products_with_most_5star = grouped.sort_values(by=["overall", "repetitions"], ascending=False)[:10][["title", "brand", "repetitions"]]
Top_10_Products_with_most_5star
```

## Top 10 Brands with Most Reviews:

```python
grouped = train_data_complete.groupby("brand").size().reset_index(name="count").sort_values(by="count", ascending=False)[:10]
grouped["average_point"] = grouped["brand"].apply(lambda X : train_data_complete.loc[train_data_complete["brand"]==X]["overall"].mean())
grouped
```

```python
grouped[["brand", "average_point"]].sort_values(by="average_point", ascending=False).reset_index(drop=True)
```

```python
from gensim.models import Word2Vec
from gensim.models import KeyedVectors
```

```python
model = KeyedVectors.load_word2vec_format('/content/drive/MyDrive/Quera Project 2/GoogleNews-vectors-negative300.bin', binary=True)
```

```python
guearantee_similar_words = model.most_similar('guarantee', topn=7)
```

```python
guearantee_similar_words
```

```python
similar_words = []
[similar_words.append(i[0]) for i in guearantee_similar_words]
```

```python
warantee_similar_words = model.most_similar('warranty', topn=10)
warantee_similar_words
```

```python
[similar_words.append(i[0]) for i in warantee_similar_words]
```
